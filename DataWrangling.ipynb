# Replacing Missing Values With The Average Of That Column

> We will replace the missing values if there is situation in which we can't bear to remove them.

```python
#taking the mean of the age column by using mean function
ks1 = sns.load_dataset('Titanic')

mean = ks1["age"].mean()
mean
ks1["age"] = ks1["age"].replace(np.nan, mean) # replacing the age null values with mean of the age column
```

# Data Normalization

Normalize data means we data ko chota kar dety hai for comparision. For example, 2 big whales kay comparision k liye we need a big space and more resources and time for management. Ye kaam hum botat easily bhi kar sakty hai by whales ko chota kar k unko ek jaar mei daal diya or ab dono ka comparision start kary gay easily:); That's called Data Normalizaton means hum data ko chota kar dety hai(screw).

Normalize data per parametric analysis lagty hai AND non-normalize data per non-parametiric or test analysis apply karty hai
- what is normalize data?
1. A data which creates bell curve is called normalize data. We create bell curve with histogram



```python
ks = sns.load_dataset("titanic")
ks.head()
```

```python
# lets compare two columns
ks3 = ks[["age", "fare"]]
ks3
```

#### **There are different methods of data normalization**
 

1. Simple Feature Scaling
    - x(new) = x(old) / x(max)
2. Min-Max Method
3. Z-Score(Standard Score) -3 -to- +3
4. Log Transformation    

- The above data is in really wide range and hard to copmare so we need to normalize it
- Normalization changes the values to the range of 0-to-1(with this both the variables will have similar influence on our models)

```python
# simple Feature Scaling 
ks3["fare"] = ks3["fare"] / ks3["fare"].max()
ks3["age"] = ks3["age"] / ks3["age"].max()

ks3.head()
```

> Difference utna e hai, just data kee range kam hogai hai

Now, see the difference in previous and above new data.

```python
# Min-Max Method
ks3["fare"] = (ks3["fare"]-ks3["fare"].min()) / (ks3["fare"].max()-ks3["fare"].min())
ks3.head()
```

```python
# Z-Score Method(Standard Method)
# Its range varies from -3 to +3. Means jo ye value deta hai uski range -3 - +3 hoti hai   
ks3["fare"] = (ks["fare"]-ks["fare"].mean()) / ks3["fare"].std()
ks3.head()
```

```python
# Log Transformation Method
ks3["fare"] = np.log(ks3["fare"])
ks3.head()
```

# Binning 
- Grouping of values into smaller number of values
- Convert numeric into categories(Jawan, Bachay, Budhay, Or age 1-16, 17-30 etd)
- To have better understanding of groups
    - Low vs Mid vs High

```python
bins = np.linspace(min(kss["age"]), max(kss["age"]), 15000)
age_groups = ["Bachay", ]
kss["age"] = pd.cut(kss["age"], bins, labels=age_groups, include_lowest=True)
kss["age"]
```
